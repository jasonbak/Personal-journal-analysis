# Analysis of a personal journal using NLTK

One of our group members, Jason, has been consistently journaling since the
start of his high school senior year (Sept 2014), producing around
283,000 words (or 350 pages of single-spaced 12 font) of unfiltered
thoughts.

This is a unique source of personal data which we can use to reveal trends
in his life. We are particularly interested in identifying trends that 
promote happier writing, which may be correlated to a happier life!

There is a non-trivial amount of processing to be done on the journals
as he has journals for different years in different file types and
formats. He has a single .docx file for 2014, 2015, and 2016; a single
.tex file for 2017; and his 2018 entries have all been in Dropbox
Paper. Each of these different file formats have a different scheme of
formatting journal entries. We need to consider a data structure to
make analysis possible across varying time frames, i.e., weeks,
months, years. At the same time, our data format needs to be 
compatible with various natural language processing libraries.

We experimented with [TextBlob](http://textblob.readthedocs.io/en/dev/),
NLTK's [VADER](http://www.nltk.org/howto/sentiment.html), and Stanford's 
[CoreNLP](https://stanfordnlp.github.io/CoreNLP/) sentiment analyzers. 
We tested all three on a random sample of our journal entries, and found
that VADER was the best at accurately reflecting our protagonist's sentiments.

We also used RNNs and LSTMs via [textgenrnn](https://github.com/minimaxir/textgenrnn) to generate Jason-like entries.

## Data source processing

We'll structure our data to facilitate analysis of varying time periods. We can use a 4d list to describe different periods of time. 

Dimensions:
- 1d - (day of month, entries from a day)
- 2d - days
- 3d - months
- 4d - years

However, the journals are in a different format for every year, so we'll have to process them separately to construct this data structure.

### 2014 - 2016: Microsoft Word
There are three .docx files for all entries from 2014, 2015, and 2016. 

Every new year, he starts on the next page and writes the year as the first line on this page. At the start of every month, he writes the name of the new month on a new line. Similarly, he writes the day of the month on a new line, followed by a colon and a space. On this same line, he inputs an entry following a `-` character.

Visually, the general .docx journal format looks like:

![](img/journal_docx_format.PNG)

#### Manual processing
The 2014, 2015, and 2016 journals were originally in a single .docx file. We manually converted these into .txt files.

#### Automatic processing
As there may be multiple entries per day, we can collect all the entries for a given day in `entries_of_day` and append (`cur_day_of_month`, `entries_of_day`) to `entries_of_month` once we move onto the next day. Similarly, when we move onto the next month, we append `entries_of_month` to `entries_of_year`.

For days that have no entries, we will still append the (`cur_day_of_month`, `entries_of_day`) tuple into `entries_of_month` but set `entries_of_day` to be empty.

````python
def processTXT(raw_txt):
    """Process a raw string (read from a .txt file) of all journal entries from
    one year into a 3d list of tuples, described as (day of month, entries from
    a day).

    Args:
        raw_txt (string): raw string of entries from a year read from a .txt
            file

    Returns 
        (3d list of tuples):
            1d - (day of month, entries from a day)
            2d - days
            3d - months
    """
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']
    # Regex to find the first entry of a day
    new_day_pattern = re.compile('^[0-9]?[0-9]:')
    # Split all raw entries by \n
    split_entries = raw_txt.splitlines()

    prev_day_of_month = 0

    entries_of_year = []
    entries_of_month = []
    entries_of_day = []
    for line in split_entries:
        if line in months:
            if entries_of_month != []:
                # Nothing in entries_of_month before first entry of new year
                # New month; add entries for the previous day and month
                entries_of_month.append((prev_day_of_month, entries_of_day))
                entries_of_year.append(entries_of_month)

                prev_day_of_month = 0
                entries_of_day = []
                entries_of_month = []
        else:
            # Every entry starts with or without '-'
            entry = line[line.find('-')+1:]

            if new_day_pattern.search(line) != None:
                # First entry of the day

                if entries_of_day != []:
                    # Nothing in entries_of_day on first entry of new month
                    # Add entries for the previous day
                    entries_of_month.append((prev_day_of_month, entries_of_day))

                # Day of month ends with a ':'
                cur_day_of_month = int(line[:line.find(':')])

                # Handle days without any entries
                for missing_day in range(prev_day_of_month+1, cur_day_of_month):
                    entries_of_month.append((missing_day, []))

                entries_of_day = [entry]
                prev_day_of_month = cur_day_of_month
            else:
                entries_of_day.append(entry)

    # Add entries for last day and month of the year
    entries_of_month.append((prev_day_of_month, entries_of_day))
    entries_of_year.append(entries_of_month)

    return entries_of_year
````

### 2017: LaTeX
He was excited about taking a discrete math course in his sophomore spring, so he wrote the 2017 journal entries in LaTeX. 

Visually, the general .tex journal format looks like:

![](img/journal_tex_format.PNG)

#### Automatic processing
In the actual entries, he only used `\textit{}` and `\textbf{}`, which our sentiment analyzer doesn't use to calculate polarity score, so we can keep that in our data.

He fills in missing days with N/A for the title and don't add anything to the actual entry, so we don't need to handle missing days in the code.

````python
def processTEX(raw_tex):
    """Process a raw string (read from a .tex file) of all journal entries from
    2017 into a 3d list of tuples, described as (day of month, entries from a
    day).

    Args:
        raw_tex (string): raw string of entries from a year read from a .tex
            file

    Returns
        (3d list of tuples):
            1d - (day of month, entries from a day)
            2d - days
            3d - months
    """
    # Ignore all the LaTeX set-up
    start_idx = raw_tex.find('\section{January}')
    end_idx = raw_tex.find('\end{document}')
    usable_str = raw_tex[start_idx:end_idx]
    # Split all entries by \n
    split_entries = usable_str.splitlines()

    day_of_month = 1
    title = ''

    titles_of_year = []
    titles_of_month = []

    entries_of_year = []
    entries_of_month = []
    entries_of_day = []
    for line in split_entries:
        if '\section' in line:
            if entries_of_month != []:
                # Nothing in entries_of_month before the entry of new year
                # New month; add entries for the previous day and month
                entries_of_month.append((day_of_month, entries_of_day))
                entries_of_year.append(entries_of_month)

                # add titles for previous month
                titles_of_year.append(titles_of_month)

                day_of_month = 1
                title = ''
                entries_of_day = []
                entries_of_month = []
                titles_of_month = []
        elif '\subsection' in line:
            if entries_of_day != []:
                # Nothing in entries_of_day on first entry of new month
                # Add entries for the previous day
                entries_of_month.append((day_of_month, entries_of_day))
                entries_of_day = []

            # Add title of day
            start_idx = len('\subsection{')
            end_idx = len(line) - 1
            title = line[start_idx:end_idx]
            titles_of_month.append(title)

            day_of_month += 1
        elif not ('%' in line and '\%' not in line):
            if line != '' or (line == '' and title == 'N/A'):
                # Fill in missing entries with ''
                # Ignore comment-only lines
                entries_of_day.append(line)

    # Add entries for last day and month of the year
    entries_of_month.append((day_of_month, entries_of_day))
    entries_of_year.append(entries_of_month)
    # Add titles for the last month of the year
    titles_of_year.append(titles_of_month)

    return entries_of_year, titles_of_year
````

### 2018: Dropbox Paper
His friend was excited for his upcoming summer internship with Dropbox and felt the desire to be a Dropbox evangelist. The friend convinced Jason to use Dropbox Paper for his journal, which he has been using since January 2018.

The format of each entry is:

![](img/journal_md_format.PNG)

#### Manual processing
Dropbox Paper is essentially an improved version of Markdown, so it can be downloaded as either a .md or .docx file. As the .md download is more accurately formatted, we will download it as a .md file.

#### Automatic processing
NLTK Vader does not process emojis but processes emoticons when analyzing sentiment. We'll replace all emojis with their emoticon equivalent.

````python
def replaceEmojis(line):
    """Replaces emojis with emoticons.

    Args:
        line (string): text content

    Returns 
        line with emoji replaced by emoticons
    """
    line = line.replace('ðŸ™‚', ':)')
    line = line.replace('ðŸ˜ž', ':(')
    line = line.replace('ðŸ˜¢', ":'(")
    line = line.replace('ðŸ˜„', ':D')
    return line
````

The newest entries are located at the start of the file, but we cannot simply reverse the order of the lines in the file since then the entries would be out of order. We need to reorder the entries within each month so that they are in chronological order, after which we reorder the month entries so they are in chronological order.

As we process the .md file, we know the month in which all the following entries are from, and because we specify the day of month in the title of the entry, we can input the text entry into the proper index, (day of month - 1), of the `entries_of_month`.

Overall processing is similar, except we populate `entries_of_month` and missing days in reverse order.

We'll ignore the meta-data for this round of analysis because there isn't enough data to provide any meaningful analysis.

````python
def processMD(raw_md):
    """Process a raw string (read from a .md file) of all journal entries from
    2018 into a 3d list of tuples, described as (day of month, entries from a
    day).

    Args:
        raw_md (string): raw string of entries from a year
            read from a .md file

    Returns
        (3d list of tuples):
            1d - (day of month, entries from a day)
            2d - days
            3d - months
    """
    months = {'January':31, 'February':28, 'March':31, 'April':30, 'May':31,
            'June':30, 'July':31, 'August':31, 'September':30, 'October':31,
            'November':30, 'December':31}
    new_month = '# '
    new_day = '## **'
    # Split all raw entries by \n
    split_entries = raw_md.splitlines()

    last_day_recorded = 1

    titles_of_year = []
    titles_of_month = []

    entries_of_year = []
    entries_of_month = []
    entries_of_day = []
    for line in split_entries:
        # NOTE: Change line below if you use another classifier that
        # handles emojis when determining sentiment
        # NTLK Vader doesn't consider emojis, but it does consider emoticons
        line = replaceEmojis(line)
        if line[:len(new_month)] == new_month and entries_of_month != []:
            month = line[len(new_month):]
            # Nothing in entries_of_month when we start processing the year
            # New month; add entries for the last day recorded and prev month
            entries_of_month.append((cur_day_of_month, entries_of_day))
            # Recorded entries in a month in reverse chronological order
            entries_of_year.append(entries_of_month[::-1])

            # Add titles for the previous month
            titles_of_year.append(titles_of_month)

            last_day_recorded = months[month]
            entries_of_day = []
            entries_of_month = []
            titles_of_month = []
        elif line[:len(new_day)] == new_day:
            # New day; on the title for entry
            if entries_of_day != []:
                # Add entries for the previous day
                entries_of_month.append((cur_day_of_month, entries_of_day))

            day_idx = line.find(':')
            cur_day_of_month = int(line[len(new_day):day_idx])

            # Handle days without any entries
            for missing_day in range(last_day_recorded-1, cur_day_of_month, -1):
                entries_of_month.append((missing_day, []))
                titles_of_month.append('N/A')

            title_idx = day_idx + len('**')
            title = line[title_idx:-len('**')]
            titles_of_month.append(title)

            entries_of_day = []
            last_day_recorded = cur_day_of_month
        elif line != '' and line[0] in string.ascii_letters+string.digits:
            # Entries only start with an alphanumeric character
            entries_of_day.append(line)

    # Add entries for last day and month of the year
    entries_of_month.append((cur_day_of_month, entries_of_day))
    # Recorded entries in a month in reverse chronological order
    entries_of_year.append(entries_of_month[::-1])

    return entries_of_year[::-1], titles_of_year[::-1]
````

````python
def processJournals():
    """Retrieve individual entries from 2014, 2015, and 2016 .txt files; 2017
    LaTeX file; and 2018 .md file. Populate a 4d list of entries and titles in
    chronological order.

    Args:
        None

    Returns
        (4d list of tuples):
            1d - (day of month, entries from a day)
            2d - days
            3d - months
            4d - years
        (4d list of strings):
            1d - title for the day
            2d - days
            3d - months
            4d - years
    """
    journal_2014 = 'data/2014.txt'
    journal_2015 = 'data/2015.txt'
    journal_2016 = 'data/2016.txt'
    journal_2017 = 'data/2017.tex'
    journal_2018 = 'data/2018.md'

    all_entries = []
    all_titles = []
    # Default mode == 'r'
    with open(journal_2014) as f14, open(journal_2015) as f15,\
        open(journal_2016) as f16, open(journal_2017) as f17,\
        open(journal_2018) as f18:
        # Process .txt files from 2014, 2015, and 2016
        raw_txt_entries_of_years = f14.read(), f15.read(), f16.read()

        for raw_txt_entries_of_year in raw_txt_entries_of_years:
            # Gets 3d list of entries
            entries_year = processTXT(raw_txt_entries_of_year)
            all_entries.append(entries_year)

        # Process .tex file from 2017
        entries_2017, titles_2017 = processTEX(f17.read())
        all_entries.append(entries_2017)
        all_titles.append(titles_2017)

        # Process .md file from 2018
        entries_2018, titles_2018 = processMD(f18.read())
        all_entries.append(entries_2018)
        all_titles.append(titles_2018)

    return all_entries, all_titles

# 4d list of entries
all_entries, all_titles = processJournals()
````

## Sentiment analysis
We'll be using NLTK's built-in Vader Sentiment Analyzer. We had success with Stanford CoreNLP, which gave us marginally better sentiment results but at the cost of drastically increasing time to run our code. Originally, we used TextBlob, whose classifier is trained on a big dataset of movie reviews from nltk.corpus. However, we found Vader to be slightly more accurate. Perhaps, Jason just doesn't write as a movie reviewer would! 

### Question: is Jason positive, negative or neutral in his writing?

We will answer this question by looking at the overall sentiment of Jason's writing across the years.

#### Overall sentiment
We've constructed our data structure so that we can grab entries from varying time periods easily. NTLK Vader requires a string to perform sentiment analysis on, so we'll create helpers to get strings of journal entries of the varying time periods that we want to analyze.

The `polarity_scores` method of the `SentimentIntensityAnalyzer` returns a dictionary with `compound`, `neg`, `neu`, and `pos` keys and values corresponding to their sentiment scores. The scores are a float within the range [0.0, 1.0], where 0.0 is not very indicative and 1.0 is very indicative.

````python
def getAllEntriesStr(all_entries):
    """Construct a string of all the entries over all the journals.

    Args:
        all_entries (4d list of tuples):
                        1d - (day of month, entries from a day)
                        2d - days
                        3d - months
                        4d - years
                        
    Returns:
        (string): a string of all the entries over all journals
    """
    all_entries_str = ''
    for year in all_entries:
        for month in year:
            for day in month:
                for entry in day[1]:
                    all_entries_str += (entry + '\n')
    return all_entries_str
````

````python
def reportOverallSentiment(all_entries):
    """Returns sentiment scores from the overall journal.

    Args:
        all_entries (4d list of tuples):
                        1d - (day of month, entries from a day)
                        2d - days
                        3d - months
                        4d - years

    Returns:
        (dict): compound, negative, neutral, and positive sentiment scores
    """
    # Condense entries over all the years into one string
    all_entries_str = getAllEntriesStr(all_entries)
    
    # This code takes 282 seconds to run
    sid = SentimentIntensityAnalyzer()
    ss = sid.polarity_scores(all_entries_str)
    return ss

overall_dict = reportOverallSentiment(all_entries)
overall_sentiment = overall_dict['pos'] - overall_dict['neg']
print('pos:', overall_dict['pos'])
print('neg:', overall_dict['neg'])
print('neu:', overall_dict['neu'])
print('overall sentiment:',overall_sentiment)
````

To measure sentiment, we will be using this simple formula:

$$sentiment = posScore-negScore$$

From 9/1/14 to 3/14/18, the sentiment of the journal has been $0.131$, which is fairly neutral. In retrospect, Jason believes he writes fairly neutrally (which can be seen through his high neutral score of $0.755$), so this is expected.

````python
def plotLineGraph(title, x_label, y_label, x_vals, y_vals, x_ticks=None, marker=True, rotation=45):
    """Helper function that wraps pyplot to easily draw line graphs.

    Args:
        title (string): title of plot
        x_label (string): x axis label
        y_label (string): y axis label
        x_vals (list): list of x values to plot
        y_vals (list): list of y values to plot
        x_ticks (list): list of x ticks
        marker (bool): True to use a circle marker
        rotation (int/string): rotation argument to xticks

    Returns 
        None
    """
    plt.figure(figsize=(15,7))

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True, which="both")
    # Make x_labels vertical
    if x_ticks != None:
        plt.xticks(x_vals, x_ticks, rotation=rotation)
    if marker == True:
        plt.plot(x_vals, y_vals, marker='o')
    else:
        plt.plot(x_vals, y_vals)
    plt.draw()
````

````python
def getEntriesOfDayStr(day):
    """Construct a string of all entries from a day

    Args:
        year (2d list of tuples):
                        1d - (day of month, entries from a day)
                        
    Returns:
        (string): a string of all the entries from one day
    """
    entries_of_day_str = ''
    for entry in day[1]:
        entries_of_day_str += (entry + '\n')
    return entries_of_day_str
````

````python
def getProperNounSentiment(all_entries):
    """Returns the sentiment of each entry keyed by the number of names that appeared in it.

    Args:
        all_entries (list): all our journal entries
        
    Returns 
        dict mapping count of names -> list of sentiment scores observed for that count
    """
    sid = SentimentIntensityAnalyzer()
    
    res = {}
    
    with open('names.txt') as f:
        names = f.read().splitlines()
        
    for year in all_entries:
        for month in year:
            for day in month:
                entries_of_day_str = getEntriesOfDayStr(day)
                num_names = len([w for w in entries_of_day_str.split(' ') if w.upper() in names])
                scores = sid.polarity_scores(entries_of_day_str)

                res[num_names] = res.get(num_names, [])
                res[num_names].append(scores['pos'] - scores['neg'])

    # to average, res = {k: sum(v)/len(v) for k,v in res.items()}
    return res

pns = getProperNounSentiment(all_entries)
````

````python
def reportTop2Sentiments(all_entries):
    """Print dates of sentiment values of the two days with the highest sentiments.

    Args:
        all_entries (4d list of tuples):
                        1d - (day of month, entries from a day)
                        2d - days
                        3d - months
                        4d - years

    Returns:
        None
    """
    top_vals = []
    top_dates = []
    top_entries = []

    sid = SentimentIntensityAnalyzer()
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']
    cur_year = 2014

    for year in all_entries:
        # Start of a new year
        # Journal starts on the 8th month in 2014
        cur_month_idx = 0 if cur_year != 14 else 8
        for month in year:
            cur_day = 1
            # Append all entries from a month into one string
            for day in month:
                entries_of_day_str = getEntriesOfDayStr(day)
                ss = sid.polarity_scores(entries_of_day_str)
                sentiment_of_day = ss['pos']-ss['neg']

                # Build up top 2
                if len(entries_of_day_str) > 2000:
                    if len(top_vals) < 2:
                        top_vals.append(sentiment_of_day)
                        cur_date = months[cur_month_idx] + ' ' \
                                + str(cur_day) + ', ' + str(cur_year)
                        top_dates.append(cur_date)
                        top_entries.append(entries_of_day_str)
                    elif min(top_vals) < sentiment_of_day:
                        min_idx = top_vals.index(min(top_vals))

                        top_vals[min_idx] = sentiment_of_day
                        cur_date = months[cur_month_idx] + ' ' \
                                + str(cur_day) + ', ' + str(cur_year)
                        top_dates[min_idx] = cur_date
                        top_entries[min_idx] = entries_of_day_str

                cur_day += 1
            cur_month_idx += 1
        cur_year += 1

    for i in range(len(top_vals)):
        print(top_dates[i] + ': ' + str(top_vals[i]))
        
reportTop2Sentiments(all_entries)
````

## Frequency analysis

### Question: what does Jason talk about?

We're interested in what the most frequent words in his journal are. This is fairly easy with TextBlob, since a TextBlob object has the same methods as a python string. We don't care about filler words like, "the", "and", etc. We can grab all the nouns of the journal using the `noun_phrases` attribute of a `TextBlob` object.

````python
def getSortedFreqs(all_words):
    """Map word to their frequencies
    
    Args:
        all_words (list): all words across all journals
        
    Returns:
       (list): list of (word, frequency) sorted by frequency in decreasing order
    """
    freq = {}
    for word in all_words:
        freq[word] = freq[word]+1 if word in freq else 1

    length = len(all_words)

    sorted_freqs = sorted(freq.items(), key=lambda x: x[1], reverse=True)

    return sorted_freqs
````

````python
def getOverallNounFrequencies(all_entries):
    """Get frequencies of nouns across all journals.
    
    Args:
        all_entries (string): all entries across all journals
        
    Returns:
        (list): list of nouns
        (list): list of frequencies
    """
    all_entries_str = getAllEntriesStr(all_entries)
    entire_journal = TextBlob(all_entries_str)

    # This line takes 20 minutes to run :(
    all_words = entire_journal.noun_phrases

    sorted_freqs = getSortedFreqs(all_words)
    return zip(*sorted_freqs)
    
top_nouns, noun_scores = getOverallNounFrequencies(all_entries)
````

### Question: what were Jason's temporary interests?
We can compare the frequencies of words from month to month to see if Jason suddenly picked up any new hobbies.

````python
def getIncreasedFrequencies(all_entries):
    """Returns a list of the five most common new words broken down per month.

    Args:
        all_entries (list): all our journal entries
        
    Returns:
        (list): the five most common new words broken down per month
    """
    
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
            'Oct', 'Nov', 'Dec']
    all_monthly_freqs = []

    for year_num, year in enumerate(all_entries, 2014):
        if year_num == 2014:
            zipped_months = zip(months[8:], year)
        else:
            zipped_months = zip(months, year)
        for month, month_data in zipped_months:
            # Append all entries from a month into one string
            entries_of_month_str = ''
            for day in month_data:
                for entry in day[1]:
                    entries_of_month_str += (entry + '\n')
            journal_of_month = TextBlob(entries_of_month_str)
            all_words = journal_of_month.words

            sorted_freqs = getSortedFreqs(all_words)
            all_monthly_freqs.append(sorted_freqs)

    top_increases = []
    
    counts = []
    for data in all_monthly_freqs:
        counts.append(Counter({k:v for (k,v) in data}))
    diffs = [x-y for (x,y) in zip(counts[1:], counts)]
    
    return [d.most_common(5) for d in diffs]

    
increased_freqs = getIncreasedFrequencies(all_entries)
````

### Positive, negative sentiment correlation

We attempted to determine the frequency of happy, sad, and neutral words per month.

````python
def reportPosNegSentiments(all_entries):
    """Graph line graph of positive, negative, and neural sentiments per entry

    Args:
        all_entries (4d list of tuples):
                        1d - (day of month, entries from a day)
                        2d - days
                        3d - months
                        4d - years

    Returns:
        None
    """
    x_labels = []

    pos = []
    neg = []
    neu = []
    sid = SentimentIntensityAnalyzer()
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
            'Oct', 'Nov', 'Dec']
    cur_year = 14
    for year in all_entries:
        # Start of a new year
        if cur_year == 14:
            cur_month_idx = 8
        else:
            cur_month_idx = 0
        for month in year:
            pos_count = 0
            neg_count = 0
            neu_count = 0
            # Build up x variables
            x_labels.append(months[cur_month_idx] + '-' + str(cur_year))
            entries_of_month_str = ''
            for day in month:
                for entry in day[1]:
                    entries_of_month_str += (entry + '\n')
            for word in entries_of_month_str.split():
                ss = sid.polarity_scores(word)
                if ss['pos'] == 1.0:
                    pos_count += 1
                elif ss['neg'] == 1.0:
                    neg_count += 1
                elif ss['neu'] == 1.0:
                    neu_count += 1
            total = pos_count + neg_count + neu_count
            pos.append(pos_count/total)
            neg.append(neg_count/total)
            neu.append(neu_count/total)
            cur_month_idx += 1
        cur_year += 1

    # Plot All sentiment types per month line graph
    x = [i for i in range(len(pos))]
    plt.figure(figsize=(11,7))
    plt.title('All sentiment types per month')
    plt.plot(x, pos, 'g')
    plt.plot(x, neg, 'r')
    plt.plot(x, neu, 'b')
    plt.xticks(x, x_labels, rotation='vertical')
    plt.ylabel('Percent')
    plt.xlabel('Month')
    plt.draw() # draw the plot
    
reportPosNegSentiments(all_entries)
````

````python
def reportYearlyWordCounts(all_entries):
    """Graph the word counts by year.

    Args:
        all_entries (4d list of tuples):
                        1d - (day of month, entries from a day)
                        2d - days
                        3d - months
                        4d - years

    Returns:
        None
    """
    # For line graph
    x_vals = ['2014', '2015', '2016', '2017', '2018']
    y_vals = []

    cur_year = 2014
    for year in all_entries:
        entries_of_year_str = getEntriesOfYearStr(year)
        # Build up y variables
        y_vals.append(len(entries_of_year_str.split()))
        cur_year += 1


    # Plot Year vs. Word count line graph
    title = 'Word count by year'
    x_label = 'Year'
    y_label = 'Word count'
    plotLineGraph(title, x_label, y_label, x_vals, y_vals)

reportYearlyWordCounts(all_entries)
````

````python
def reportMonthlyWordCounts(all_entries):
    """Graph the word counts by month.

    Args:
        all_entries (4d list of tuples):
                        1d - (day of month, entries from a day)
                        2d - days
                        3d - months
                        4d - years

    Returns:
        None
    """
    # For line graph
    x_ticks = []
    y_vals = []

    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
            'Oct', 'Nov', 'Dec']
    cur_year = 14
    for year in all_entries:
        # Start of a new year
        if cur_year == 14:
            cur_month_idx = 8
        else:
            cur_month_idx = 0
        for month in year:
            # Build up x variables
            x_ticks.append(months[cur_month_idx] + '-' + str(cur_year))
            entries_of_month_str = getEntriesOfMonthStr(month)

            # Build up y variables
            y_vals.append(len(entries_of_month_str.split()))
            cur_month_idx += 1
        cur_year += 1

    # Plot Month vs. Word count line graph
    title = 'Word count by month'
    x_label = 'Month'
    y_label = 'Word count'
    x_vals = [i for i in range(len(y_vals))]
    plotLineGraph(title, x_label, y_label, x_vals, y_vals, x_ticks=x_ticks)

reportMonthlyWordCounts(all_entries)
````
